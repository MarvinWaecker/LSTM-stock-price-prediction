## Arbeitspaket 2: Datenaufbereitung

Nun wird die Datenaufbereitung für das maschinelle Lernen durchgeführt. Dabei ist die Zielvariable der Closing-Kurs der Aktie und die Features die Spalten 'Date', 'Open', 'High', 'Low‘. Für eine Vereinfachung wird das Feature Volumen weggelassen.

#### Load CSV

Methode aus Arbeitspaket 1 erneut aufrufen:

```python
def read_csv(stock_name):
    df = pd.read_csv(stock_name, index_col = 0)
    # Feature Volume entfernen
    df.drop(['Volume'], 1, inplace=True)
    # Feature Adj. Close entfernen
    df.drop(['Adj Close'], 1, inplace=True)
    return df
```

#### Normalize Data

Durch die Normalisierung der Daten wird das Modell besser trainiert und alle Daten werden so skaliert, dass sie zwischen 1 und -1 liegen.
Wir haben diesen Code als Methode verpackt. Dies macht Sinn, da dieser mehrmals gebraucht wird.

```python
def normalize_data(df):
    min_max_scaler = MinMaxScaler()
    df['Open'] = min_max_scaler.fit_transform(df.Open.values.reshape(-1,1))
    df['High'] = min_max_scaler.fit_transform(df.High.values.reshape(-1,1))
    df['Low'] = min_max_scaler.fit_transform(df.Low.values.reshape(-1,1))
    df['Close'] = min_max_scaler.fit_transform(df.Close.values.reshape(-1,1))
  
    return df
```

### Load Data

Ein neuronales Netz wird immer in eine Trainings- und Testmenge aufgeteilt. 

Für die Form der Trainingsmenge gibt es prinzipiell 2 Möglichkeiten:

**1. Möglichkeit:** 
Für die Trainingsmenge werden die Daten bis zu einem gewissen Zeitpunkt genommen, bspw. von Veröffentlichung der Aktie bis hin zum 01.01.2020 und nutzen für die Testmenge alle Daten ab 01.01.2020

**2. Möglichkeit:** 
Sliding Window: Bilden eines Fensters, bspw. 18 Tage, das entlang des gesamten Zeitraums der Aktie geschoben wird. Dabei werden die ersten 17 Tage der Sequenz genutzt, um den Close Price des nächsten Tages vorherzusagen.


Wir haben uns für die zweite Möglichkeit entschieden. Durch das Verfahren 'Sliding Window' wird ein Fenster mit 18 Tagen für die Aktienkurse Apple und AT&T erfasst. Diese werden in 17 Tage Trainingsmenge und 15 Tage Testmenge aufgeteilt. Dies entspricht einer Aufteilung der Daten in 75 Prozent Trainingsmenge und 25 Prozent Testmenge.

```python
def load_data(stock, seq_len): 
    
    amount_of_features = len(stock.columns)
    #data = stock.as_matrix()
    data = stock.values

    
    features = []
    labels = []
    
    for i in range(seq_len,len(data)): # 13, 1259
        features.append(data[i-seq_len:i]) # data(0-13:0)
        labels.append(data[i,-1])
        #X_train.append(training_set_scaled[i-60:i, 0]) # was 60
    
    features = np.array(features)
    labels = np.array(labels)
    row = int(round(0.8 * features.shape[0])) # 994
    row_val = int(round(0.8 * row)) # 795
    # print(features.shape[0]) # 1243
  
    
    x_train = features[:row_val] 
    y_train = labels[:row_val]
    
    x_valid = features[row_val:row]
    y_valid = labels[row_val:row]
    
    x_test = features[row:]
    y_test = labels[row:]
    
    # train shape (795, 13, 4)
    # valid shape (199, 13, 4)
    # test shape (249, 13, 4)

    # x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], amount_of_features))
    # x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], amount_of_features))  
    return [x_train, y_train, x_valid, y_valid, x_test, y_test]
```

### Denormalize Data

Um die normalisierten Werte zwischen -1 und 1 wieder auf Aktienwerte zurückzuführen, werden die Daten denormalisiert.

```python
def denormalize(df, normalized_value): 
    min_max_scaler = MinMaxScaler()
    df['Close'] = min_max_scaler.fit_transform(df.Close.values.reshape(-1,1))
    new = min_max_scaler.inverse_transform(normalized_value.reshape(-1,1))
    return new
```

### Evaluate Model

Das  Modell muss entsprechend evaluiert werden, um nachvollziehen zu können wie gut es trainiert bzw. die Testdaten vorhergesagt hat. Dies geschieht über den Mean Squarred Error.

```python
def model_score(model, X_train, X_valid, y_valid, y_train, X_test, y_test):
    trainScore = model.evaluate(X_train, y_train, verbose=0)
    print('Train Score: %.5f MSE (%.2f RMSE)' % (trainScore[0], math.sqrt(trainScore[0])))
    
    validScore = model.evaluate(X_valid, y_valid, verbose=0)
    print('Validation Score: %.5f MSE (%.2f RMSE)' % (validScore[0], math.sqrt(validScore[0])))

    testScore = model.evaluate(X_test, y_test, verbose=0)
    print('Test Score: %.5f MSE (%.2f RMSE)' % (testScore[0], math.sqrt(testScore[0])))
```

**Fragen zu den beiden Möglichkeiten/Verfahren:** 

(Vorbereitung für die nächsten Arbeitspakete)

**Was eignet sich für Möglichkeit 1?**

- Konstant wachsende Aktien über einen langen Zeitraum
- Vorteil: In der Gesamtheit können bei konstant wachsenden Aktien bessere und langfristige Ergebnisse vorausgesagt werden.
- Nachteil: Es können keine kurzfristigen Aussagen getroffen werden, bspw. aufgrund aktueller Geschehnisse die den Aktienkurs stark beeinflussen.


**Was eignet sich für Möglichkeit 2 (Sliding Window)?**

- (stark) schwankende Aktien

- Vorteil: Bei schwankenden Aktien wird ein kürzerer Zeitraum  betrachtet und dadurch kann eine genauere Prognose getroffen werden. Es wird der maximale Gewinn aus den Daten geschöpft, die gegeben sind, um immer wieder neue Vorhersagen zu treffen.
- Nachteil: Es sind keine langfristigen Aussagen möglich.

**Welches Verfahren eignet sich am besten für BionTech, Moderna & Crispr?**

Antwort: Sliding Window

- **BionTech:** Bei BionTech kam es seit März 2020 zu einer unvorhersehbaren Aktiensteigerung in den letzten 6 Monaten. Die Zeit davor ist für die Betrachtung des Aktienkurses eher irrelevant.  Es ist wichtig die unmittelbare Zukunft von BionTech zu betrachten, deswegen ist das Sliding Window besser. Informationen werden regelmäßig geupdatet und fließen in das neue Modell mit ein.
- **Moderna:** siehe BionTech
- **Crispr:** Sind erst seit kurzem an der Börse (2016), entsprechend gibt es weniger Daten und ebenfalls wie bei BionTech ist ein Sliding Window besser geeignet

### Load data

Die Einheit 16 für das Window entspricht 60 Tagen in Paketen. Dies wird für das Dataset der Apple-Aktie vorgenommen.
Es wird außerdem untersucht, ob die Aufteilung in die Trainings- und Testdaten geklappt hat. Wenn im Training die gleiche Anzahl an Zeilen wäre wie im normalen Data Frame, hätte es nicht geklappt.



```python
window = 13

X_train, y_train, X_valid, y_valid, X_test, y_test = load_data(df_apple, window)

# print(len(df_apple)) # Anzahl der Observationen im Apple Datensatz
# print (X_train.shape)
# print(y_train)# Dimension des Trainingsdatensatzes

window = len(X_train)//60 # entspricht 60 Tage Pakete
```

## Arbeitspaket 3: RNN

RNN = Rekurrentes neuronales Netz

Durch die rekurrente Schicht können RNN-Modelle Informationen im Laufe der Zeit im "Gedächtnis" behalten. Jedoch zeigen sich Schwierigkeiten bei Datensets, die die Betrachtung langfristiger zeitlicher Abhängigkeiten erfordern. Das bedeutet, dass RNN's nicht dafür geeignet sind, um sich längere/genauere Muster über einen langen Zeitraum zu merken.
Im Hinblick auf die Aufgabe, lässt sich somit annehmen, dass das RNN-Modell ungenaue Vorhersagen trifft.

### Build Model

Hier wird das Netzwerk gebildet und daraufhin das entsprechende Window gesetzt.

```python
def build_model_rnn(layers):
    d = 0.3
    model = Sequential()
    
    # LSTM durch SimpleRNN ersetzt
    model.add(SimpleRNN(256, input_shape=(layers[0], layers[1]), return_sequences=True))
    model.add(Dropout(d))
    
    # LSTM durch SimpleRNN ersetzt
    model.add(SimpleRNN(256, return_sequences=False))
    model.add(Dropout(d))
        
    model.add(Dense(32,kernel_initializer="uniform",activation='relu'))        
    model.add(Dense(1,kernel_initializer="uniform",activation='linear'))

        
    start = time.time()
    model.compile(loss='mse',optimizer='adam', metrics=['mean_squared_error'])
    print("Compilation Time : ", time.time() - start)
    return model
```

Die Compilation Time bezieht sich auf die Zeitdauer, in der der Programmiercode in den Machine Learning Code umgewandelt wird.

```python
model_rnn = build_model_rnn([window,X_train.shape[2]])
```

### Fit Model

Nun wird das Modell trainiert. Die Epochen stellen dabei die Durchläufe über den gesamten Trainingsdatensatz dar.

```python
model_rnn.fit(X_train,y_train,batch_size=256,epochs=90,validation_data=(X_valid, y_valid),verbose=1)
```

Nun werden die Differenzen und das Verhältnis zwischen den wahren und vorhergesagten Y-Werten für die Testdaten berechnet.

```python
p_rnn = model_rnn.predict(X_test)
print(p_rnn.shape)
```

### Denormalize Data

```python
df_apple_fit_scaler = pd.read_csv('AAPL.csv')
newp_rnn = denormalize(df_apple_fit_scaler, p_rnn)
df_apple_fit_scaler = pd.read_csv('AAPL.csv')
newy_test = denormalize(df_apple_fit_scaler, y_test)
```

### Evaluate Data

Der Mean Squarred Error gibt die mittlere quadratische Abweichung der Trainings- und Testmenge an. Anhand dessen kann evaluiert werden, wie genau das trainierte Modell auf Trainings- und Testdaten Vorhersagen trifft.

```python
model_score(model_rnn, X_train, X_valid, y_valid, y_train, X_test, y_test)
```

## RNN Tuning

Das RNN-Netzwerk wird jetzt noch einmal variiert bzw. getuned, um dadurch mögliche Änderungen bezüglich der Genauigkeit zum vorherigen Netzwerk aufzuzeigen. Durch das Tuning werden dem Algorithmus Spezifikationen gegeben, womit dieser besser lernen kann. Dabei werden die gleichen Schritte wie beim vorherigen RNN-Modell verfolgt, jedoch verschiedene Faktoren innerhalb der Programmiercodes geändert.

Tuning anhand Veränderung der:

- Anzahl Layer
- Parameter
- Anzahl Trainingsdurchläufe

### Build Model

Änderung der Anzahl an Layern, Anzahl an Neuronen auf 50 und der Dropout-Rate auf 0.2.

```python
def build_model_rnn_tuned(layers):
    d = 0.2
    model = Sequential()
    
    model.add(SimpleRNN(50, input_shape=(layers[1], layers[0]), return_sequences=True))
    model.add(Dropout(d))
        
    model.add(SimpleRNN(50, input_shape=(layers[1], layers[0]), return_sequences=True))
    model.add(Dropout(d))
    
    model.add(SimpleRNN(50, input_shape=(layers[1], layers[0]), return_sequences=True))
    model.add(Dropout(d))
    
    model.add(SimpleRNN(50, input_shape=(layers[1], layers[0])))
    model.add(Dropout(d))
              
    model.add(Dense(1))
        
    start = time.time()
    model.compile(loss='mse',optimizer='adam', metrics=['mean_squared_error'])
    print("Compilation Time : ", time.time() - start)
    return model
```

```python
model_rnn_tuned = build_model_rnn_tuned([4,window,1])
```

### Fit Model

Änderung der Trainingsdurchläufe auf 100 und batch size auf 32.

```python
model_rnn_tuned.fit(X_train,y_train,batch_size=32,epochs=100,validation_data=(X_valid, y_valid),verbose=1)
```

Nun werden die Differenzen und das Verhältnis zwischen den wahren und vorhergesagten Y-Werten für die Testdaten berechnet.

```python
p_rnn_tuned = model_rnn_tuned.predict(X_test)
print (p_rnn_tuned.shape)
```

### Denormalize Data

```python
df_apple_fit_scaler = pd.read_csv('AAPL.csv')
newp_rnn_tuned = denormalize(df_apple_fit_scaler, p_rnn_tuned)
df_apple_fit_scaler = pd.read_csv('AAPL.csv')
newy_test = denormalize(df_apple_fit_scaler, y_test)
```

### Evaluate Data

```python
model_score(model_rnn_tuned, X_train, X_valid, y_valid, y_train, X_test, y_test)
```

Im Vergleich zum vorherigen Modell lässt sich erkennen, dass die Abweichung geringer ist und das Modell durch die Änderungen genauer wird.

### Visualization


Durch die folgende Visualisierung kann graphisch dargelegt werden, inwiefern das RNN-Modell mit den Änderungen und damit vorhergesagten Werten dem aktuellen Aktienkurs entspricht.

```python
plt.figure(figsize=(15, 7));

plt.plot(newp_rnn_tuned, color='blue', label='Prediction')
plt.plot(newy_test, color='green', label='Actual')

# stats
plt.title('Prediction and actual Price Comparison')
plt.xlabel('time [days]')
plt.ylabel('price [USD]')
plt.legend(loc='best')

# show plot
plt.show()
```

Noch besser zeigt der graphische Vergleich zwischen dem ersten RNN-Modell, veränderten RNN-Modell und den aktuellen Werten die Unterschiede auf.

```python
plt.figure(figsize=(15, 7));

plt.plot(newp_rnn, color='red', label='Prediction')
plt.plot(newp_rnn_tuned, color='blue', label='Prediction Tuned')
plt.plot(newy_test, color='green', label='Actual')

# stats
plt.title('Standard vs. Tuned')
plt.xlabel('time [days]')
plt.ylabel('price [USD]')
plt.legend(loc='best')
```

## Arbeitspaket 4: LSTM






LSTM = Long-Short Term Memory Netzwerk

Das LSTM Netzwerk stellt eine Weiterführung des RNN-Modells dar, wodurch dessen Speicher im Wesentlichen erweitert wird. Es speichert Informationen und merkt sich Werte über einen langen Zeitraum besser und genauer, welche in die Entscheidungsfindung miteinbezogen werden.


Grundsätzlich lassen sich alle Programmierschritte aus dem RNN-Netzwerk mit der jeweiligen Abänderung auf das LSTM-Netzwerk übertragen.

### Build Model

```python
def build_model_lstm(layers):
    d = 0.3
    model = Sequential()
    
    model.add(LSTM(256, input_shape=(layers[1], layers[0]), return_sequences=True))
    model.add(Dropout(d))
        
    model.add(LSTM(256, input_shape=(layers[1], layers[0]), return_sequences=False))
    model.add(Dropout(d))
        
    model.add(Dense(32,kernel_initializer="uniform",activation='relu'))        
    model.add(Dense(1,kernel_initializer="uniform",activation='linear'))

        
    start = time.time()
    model.compile(loss='mse',optimizer='adam', metrics=['mean_squared_error'])
    print("Compilation Time : ", time.time() - start)
    return model
```

```python
model_lstm = build_model_lstm([4,window,1]) 
```

### Fit Model

```python
model_lstm.fit(X_train,y_train,batch_size=512,epochs=90,validation_data=(X_valid, y_valid),verbose=1)
```

Nun werden die Differenzen und das Verhältnis zwischen den wahren und vorhergesagten Y-Werten für die Testdaten berechnet.

```python
p_lstm = model_lstm.predict(X_test)
print (p_lstm.shape)
```

### Denormalize Data

```python
df_apple_fit_scaler = pd.read_csv('AAPL.csv')
newp_lstm = denormalize(df_apple_fit_scaler, p_lstm)
df_apple_fit_scaler = pd.read_csv('AAPL.csv')
newy_test = denormalize(df_apple_fit_scaler, y_test)
```

### Evaluate Model

```python
model_score(model_lstm, X_train, X_valid, y_valid, y_train, X_test, y_test)
```

### Visualization

```python
plt.figure(figsize=(15, 7));

plt.plot(newp_lstm, color='red', label='Prediction')
plt.plot(newy_test, color='green', label='Actual')

# stats
plt.title('Prediction and actual Price Comparison')
plt.xlabel('time [days]')
plt.ylabel('price [USD]')
plt.legend(loc='best')

# show plot
plt.show()
```

## Arbeitspaket 3: RNN

RNN = Rekurrentes neuronales Netz

Durch die rekurrente Schicht können RNN-Modelle Informationen im Laufe der Zeit im "Gedächtnis" behalten. Jedoch zeigen sich Schwierigkeiten bei Datensets, die die Betrachtung langfristiger zeitlicher Abhängigkeiten erfordern. Das bedeutet, dass RNN's nicht dafür geeignet sind, um sich längere/genauere Muster über einen langen Zeitraum zu merken.
Im Hinblick auf die Aufgabe, lässt sich somit annehmen, dass das RNN-Modell ungenaue Vorhersagen trifft.

### Build Model

Hier wird das Netzwerk gebildet und daraufhin das entsprechende Window gesetzt.

```python
def build_model_rnn(layers):
    d = 0.3
    model = Sequential()
    
    # LSTM durch SimpleRNN ersetzt
    model.add(SimpleRNN(256, input_shape=(layers[0], layers[1]), return_sequences=True))
    model.add(Dropout(d))
    
    # LSTM durch SimpleRNN ersetzt
    model.add(SimpleRNN(256, return_sequences=False))
    model.add(Dropout(d))
        
    model.add(Dense(32,kernel_initializer="uniform",activation='relu'))        
    model.add(Dense(1,kernel_initializer="uniform",activation='linear'))

        
    start = time.time()
    model.compile(loss='mse',optimizer='adam', metrics=['mean_squared_error'])
    print("Compilation Time : ", time.time() - start)
    return model
```

Die Compilation Time bezieht sich auf die Zeitdauer, in der der Programmiercode in den Machine Learning Code umgewandelt wird.

```python
model_rnn = build_model_rnn([window,X_train.shape[2]])
```

### Fit Model

Nun wird das Modell trainiert. Die Epochen stellen dabei die Durchläufe über den gesamten Trainingsdatensatz dar.

```python
model_rnn.fit(X_train,y_train,batch_size=256,epochs=90,validation_data=(X_valid, y_valid),verbose=1)
```

Nun werden die Differenzen und das Verhältnis zwischen den wahren und vorhergesagten Y-Werten für die Testdaten berechnet.

```python
p_rnn = model_rnn.predict(X_test)
print(p_rnn.shape)
```

### Denormalize Data

```python
df_att_fit_scaler = pd.read_csv('T.csv')
newp_rnn = denormalize(df_att_fit_scaler, p_rnn)
df_att_fit_scaler = pd.read_csv('T.csv')
newy_test = denormalize(df_att_fit_scaler, y_test)
```

### Evaluate Data

Der Mean Squarred Error gibt die mittlere quadratische Abweichung der Trainings- und Testmenge an. Anhand dessen kann evaluiert werden, wie genau das trainierte Modell auf Trainings- und Testdaten Vorhersagen trifft.

```python
model_score(model_rnn, X_train, X_valid, y_valid, y_train, X_test, y_test)
```

### Visualization


Mit einem Plot wird ein Vergleich zwischen den vorhergesagten Werten durch das Modell und den aktuellen Werten dargestellt.

```python
plt.figure(figsize=(15, 7));

plt.plot(newp_rnn, color='red', label='Prediction')
plt.plot(newy_test, color='green', label='Actual')

# stats
plt.title('Prediction and actual Price Comparison')
plt.xlabel('time [days]')
plt.ylabel('price [USD]')
plt.legend(loc='best')

# show plot
plt.show()
```

Anhand der Prediction and Acutal Price

## RNN Tuning

Das RNN-Netzwerk wird jetzt noch einmal variiert bzw. getuned, um dadurch mögliche Änderungen bezüglich der Genauigkeit zum vorherigen Netzwerk aufzuzeigen. Durch das Tuning werden dem Algorithmus Spezifikationen gegeben, womit dieser besser lernen kann. Dabei werden die gleichen Schritte wie beim vorherigen RNN-Modell verfolgt, jedoch verschiedene Faktoren innerhalb der Programmiercodes geändert.

Tuning anhand Veränderung der:

- Anzahl Layer
- Parameter
- Anzahl Trainingsdurchläufe

### Build Model

Änderung der Anzahl an Layern, Anzahl an Neuronen auf 50 und der Dropout-Rate auf 0.2.

```python
def build_model_rnn_tuned(layers):
    d = 0.2
    model = Sequential()
    
    model.add(SimpleRNN(50, input_shape=(layers[1], layers[0]), return_sequences=True))
    model.add(Dropout(d))
        
    model.add(SimpleRNN(50, input_shape=(layers[1], layers[0]), return_sequences=True))
    model.add(Dropout(d))
    
    model.add(SimpleRNN(50, input_shape=(layers[1], layers[0]), return_sequences=True))
    model.add(Dropout(d))
    
    model.add(SimpleRNN(50, input_shape=(layers[1], layers[0])))
    model.add(Dropout(d))
              
    model.add(Dense(1))
        
    start = time.time()
    model.compile(loss='mse',optimizer='adam', metrics=['mean_squared_error'])
    print("Compilation Time : ", time.time() - start)
    return model
```

```python
model_rnn_tuned = build_model_rnn_tuned([4,window,1])
```

### Fit Model

Änderung der Trainingsdurchläufe auf 100 und batch size auf 32.

```python
model_rnn_tuned.fit(X_train,y_train,batch_size=32,epochs=100,validation_data=(X_valid, y_valid),verbose=1)
```

Nun werden die Differenzen und das Verhältnis zwischen den wahren und vorhergesagten Y-Werten für die Testdaten berechnet.

```python
p_rnn_tuned = model_rnn_tuned.predict(X_test)
print (p_rnn_tuned.shape)
```

### Denormalize Data

```python
df_att_fit_scaler = pd.read_csv('T.csv')
newp_rnn_tuned = denormalize(df_att_fit_scaler, p_rnn_tuned)
df_att_fit_scaler = pd.read_csv('T.csv')
newy_test = denormalize(df_att_fit_scaler, y_test)
```

### Evaluate Data

```python
model_score(model_rnn_tuned, X_train, X_valid, y_valid, y_train, X_test, y_test)
```

Im Vergleich zum vorherigen Modell lässt sich erkennen, dass die Abweichung geringer ist und das Modell durch die Änderungen genauer wird.

### Visualization


Durch die folgende Visualisierung kann graphisch dargelegt werden, inwiefern das RNN-Modell mit den Änderungen und damit vorhergesagten Werten dem aktuellen Aktienkurs entspricht.
